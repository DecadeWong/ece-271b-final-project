# Imports & Miscellaneous Functions

```python
import os
import time
import torch
import pickle
import matplotlib.pyplot as plt
import numpy as np
from torch.utils.data import DataLoader
from torch.utils.data import TensorDataset
from torch import nn
import torch.nn.functional as F
from torch.autograd import Variable
import torch.optim.lr_scheduler


#use_cuda = False
use_cuda = torch.cuda.is_available()


def split_dataset(data, label, split_ratio = 0.8, shuffle = True):
    # create training and validation dataset
    all_idx = np.arange(len(label))
    if shuffle: 
        np.random.shuffle(all_idx)
    idx_split = int(len(label)*split_ratio)
    train_idx, valid_idx = all_idx[:idx_split], all_idx[idx_split:]
    train_data = data[train_idx]
    train_label = label[train_idx]
    valid_data = data[valid_idx]
    valid_label = label[valid_idx]
    return train_data, train_label, valid_data, valid_label


def print_bar(ratio):
    # print bar to show current progress
    print_str = '['
    progress_count = int(20.0*ratio)
    for i in range(progress_count - 1):
        print_str += '='
    if progress_count == 20:
        print_str += '='
    elif progress_count > 0:
        print_str += '>'
    for i in range(20 - progress_count):
        print_str += ' '
    print_str += ']'
    print(print_str + '\r', end = '')
```

# Read Data & Create Dataset

```python
pickle_path = os.getcwd() + '/dataset/data_pre/'
with open(pickle_path + 'lookup.pkl', 'rb') as f:
    lookup_cmd2vec, lookup_cmd2num, lookup_vec2cmd = pickle.load(f)
    f.close()
uni_cmd = len(lookup_cmd2vec)

pkl_files = [y for y in os.listdir(pickle_path) if '.pickle' in y]
data = np.zeros((1, 99,161))
lbls = np.zeros(1)
for item in pkl_files:
    with open(pickle_path + item, 'rb') as f:
        data_tmp, lbls_tmp = pickle.load(f)
        f.close()
    data = np.vstack((data, data_tmp))
    lbls = np.hstack((lbls, lbls_tmp))
data = np.expand_dims(data[1:], axis = 1)
lbls = lbls[1:]

train_data, train_lbls, test_data, test_lbls = split_dataset(data, lbls)
```

# Network & Training functions

```python
# Network
class ConvRes(nn.Module):
    def __init__(self, insize, outsize):
        super(ConvRes, self).__init__()
        self.math = nn.Sequential(
            nn.BatchNorm2d(insize),
#            nn.Dropout(p = .3),
            torch.nn.Conv2d(insize, outsize, kernel_size=2, padding=2),
            nn.PReLU(),
        )

    def forward(self, x):
        return self.math(x)


class ConvCNN(nn.Module):
    def __init__(self, insize, outsize, kernel_size=7, padding=2, pool=2, avg=True):
        super(ConvCNN, self).__init__()
        self.avg = avg
        self.math1 = torch.nn.Sequential(
            torch.nn.Conv2d(insize, outsize, kernel_size=kernel_size, padding=padding),
            torch.nn.BatchNorm2d(outsize),
            torch.nn.LeakyReLU(),
            torch.nn.MaxPool2d(pool, pool),
        )
        self.math2 = torch.nn.Sequential(
            torch.nn.Conv2d(insize, outsize, kernel_size=kernel_size, padding=padding),
            torch.nn.BatchNorm2d(outsize),
            torch.nn.LeakyReLU(),
            torch.nn.MaxPool2d(pool, pool),
        )
        self.avgpool = torch.nn.AvgPool2d(pool, pool)

    def forward(self, x):
        x1 = self.math1(x)
        x2 = self.math2(x)
        x =x1 * F.sigmoid(x2) 
        if self.avg is True:
            x = self.avgpool(x)
        return x


class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.dropout = torch.nn.Dropout(p = 0.30)
        self.avgpool = nn.AdaptiveAvgPool2d(1)
        self.cnn1 = ConvCNN( 1, 32, kernel_size=7, pool=4, avg=False)
        self.cnn2 = ConvCNN(32, 32, kernel_size=5, pool=2, avg=True)
        self.cnn3 = ConvCNN(32, 32, kernel_size=5, pool=2, avg=True)
        self.res1 = ConvRes(32, 64)

        self.features = nn.Sequential(
            self.cnn1, self.dropout,
            self.cnn2,
            self.cnn3,
            self.res1,
        )
        self.classifier = torch.nn.Sequential(
            nn.Linear(1280, uni_cmd),
        )
  
    def forward(self, x):
        x = self.features(x)
#         print (x.data.shape)
        x = self.classifier(x.view(x.size(0), -1))
        return x

    
# Training functions
def train(train_data, train_lbls, model, criterion, optimizer, batch_size):
#    model.train()
    len_train = len(train_lbls)
    print("Training:")
    print_bar(0)
    # batch for training
    batch_num = len_train // batch_size
    for batch in range(batch_num):
        print_bar((batch+1)*1.0/batch_num)
        
        loader_data = train_data[batch:batch + batch_size]
        inputs = np.reshape(loader_data, ((batch_size,) + loader_data[0].shape))
        labels = train_lbls[batch:batch + batch_size]
        # train
        if use_cuda: 
            inputs = Variable(torch.from_numpy(inputs).float().cuda())
            labels = Variable(torch.from_numpy(labels).long().cuda())
        else:
            inputs = Variable(torch.from_numpy(inputs).float())
            labels = Variable(torch.from_numpy(labels).long())
            
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
    

def compute_loss_error(data, lbls, model, criterion, batch_size = 500):
    # batch for holdout
    len_data = len(lbls)
    batch_num = len_data // batch_size
    total_loss = 0     # total loss
    total_error = 0    # total error
    print_bar(0)
    for batch in range(batch_num):
        print_bar((batch+1)*1.0/batch_num)
        
        loader_data = train_data[batch:batch + batch_size]
        inputs = np.reshape(loader_data, ((batch_size,) + loader_data[0].shape))
        labels = train_lbls[batch:batch + batch_size]
        # train
        if use_cuda: 
            inputs = Variable(torch.from_numpy(inputs).float().cuda())
            labels = Variable(torch.from_numpy(labels).long().cuda())
        else:
            inputs = Variable(torch.from_numpy(inputs).float())
            labels = Variable(torch.from_numpy(labels).long())
            
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        total_loss += loss.data.cpu().numpy()   # update total loss
        _, predicted = torch.max(outputs.data, 1)
        total_error = (total_error*batch + (predicted != labels.data).float().mean()*100)/(batch + 1)
    return total_loss*1.0/batch_num, total_error



def run_epochs(model, train_data, train_lbls, tst_data, tst_lbls, \
               criterion, optimizer, epoch_max = 100, batch_size = 128):
    loss_trn_log = np.zeros(epoch_max + 1)
    loss_vld_log = np.zeros(epoch_max + 1)
    loss_tst_log = np.zeros(epoch_max + 1)
    error_trn_log = np.zeros(epoch_max + 1)
    error_vld_log = np.zeros(epoch_max + 1)
    error_tst_log = np.zeros(epoch_max + 1)
    print('Initial status:')  
    trn_data, trn_lbls, vld_data, vld_lbls = split_dataset(train_data, train_lbls)
    loss_trn_log[0], error_trn_log[0] = \
        compute_loss_error(trn_data, trn_lbls, model, criterion, batch_size)
    loss_vld_log[0], error_vld_log[0] = \
        compute_loss_error(vld_data, vld_lbls, model, criterion, batch_size)
    loss_tst_log[0], error_tst_log[0] = \
        compute_loss_error(tst_data, tst_lbls, model, criterion, batch_size)
    print('\nTrain loss: %.3f, Train error: %.2f%%' % \
          (loss_trn_log[0], error_trn_log[0]))
    print('Valid loss: %.3f, Valid error: %.2f%%' % \
          (loss_vld_log[0], error_vld_log[0]))
    print('Test  loss: %.3f, Test  error: %.2f%%' % \
          (loss_tst_log[0], error_tst_log[0]))
    
    print('\nStart training.')
    print('---------------------')
    start_time = time.time()
    
    for epoch in range(epoch_max):
        t_epoch = time.time()
        print('EPOCH %3d:' % (epoch + 1))
        trn_data, trn_lbls, vld_data, vld_lbls = split_dataset(train_data, train_lbls)
        #train
        train(trn_data, trn_lbls, model, criterion, optimizer, batch_size)
        #calculate loss
        print("\nComputing loss & error:")
        loss_trn_log[epoch + 1], error_trn_log[epoch + 1] = \
            compute_loss_error(trn_data, trn_lbls, model, criterion, batch_size)
        loss_vld_log[epoch + 1], error_vld_log[epoch + 1] = \
            compute_loss_error(vld_data, vld_lbls, model, criterion, batch_size)
        loss_tst_log[epoch + 1], error_tst_log[epoch + 1] = \
            compute_loss_error(tst_data, tst_lbls, model, criterion, batch_size)
        print('\nTrain loss: %.3f, Train error: %.2f%%' % \
              (loss_trn_log[epoch + 1], error_trn_log[epoch + 1]))
        print('Valid loss: %.3f, Valid error: %.2f%%' % \
              (loss_vld_log[epoch + 1], error_vld_log[epoch + 1]))
        print('Test  loss: %.3f, Test  error: %.2f%%' % \
              (loss_tst_log[epoch + 1], error_tst_log[epoch + 1]))
        print('Time spent: %.2f s' % (time.time() - t_epoch))
        print('---------------------')
        scheduler.step()
    print('\nTraining ends here. Time usage: %.2f s.' % (time.time() - start_time))
    return loss_trn_log, loss_vld_log, loss_tst_log, error_trn_log, error_vld_log, error_tst_log


def plot_figure(train_loss, valid_loss, test_loss, ylabel, title = None, filename = None):
    # plot loss of training, validation and test at the end of training
    plt.figure()
    plt.plot(range(len(train_loss)), train_loss, label = 'Train')
    plt.plot(range(len(valid_loss)), valid_loss, label = 'Validation')
    plt.plot(range(len(test_loss)), valid_loss, label = 'Test')
    if title != None:
        plt.title(title)
    plt.legend()
    plt.xlabel('epochs')
    plt.ylabel(ylabel)
    plt.grid()
    if filename != None:
        plt.savefig(filename)
    plt.show()
```

# Training model

```python
# Initialize model
if use_cuda:
    print("Using the GPU.")
    model = Net().cuda()
    criterion = nn.CrossEntropyLoss().cuda()
else:
    print("Using the CPU.")
    model = Net()
    criterion = nn.CrossEntropyLoss()

lr0 = 1e-2
#optimizer = torch.optim.SGD(model.parameters(), lr = lr0, momentum = 0.5)
optimizer = torch.optim.Adam(model.parameters(), lr = lr0, weight_decay=5e-5) #  L2 regularization
lambda_lr = lambda epoch: max(.5, 0.995 ** epoch)
scheduler = torch.optim.lr_scheduler.LambdaLR(optimizer, lr_lambda = lambda_lr)

# Start training
loss_trn, loss_vld, loss_tst, error_trn, error_vld, error_tst = \
    run_epochs(model, train_data, train_lbls, test_data, test_lbls, \
               criterion, optimizer, epoch_max = 200, batch_size = 16)
# loss figure
plot_figure(loss_trn, loss_vld, loss_tst, \
            ylabel = 'loss', title = None, filename = 'loss.eps')
plot_figure(error_trn, error_vld, error_tst, \
            ylabel = 'error (%)', title = None, filename = 'error.eps')
```
